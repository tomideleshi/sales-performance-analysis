```python
#Install and import all libraries we need
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("All libraries loaded successfully!")
```

```python
# Setting a seed so our random data is consistent
np.random.seed(42)

# Creating our dataset
data = {
    'Month': ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'] * 10,
    'Salesperson': np.random.choice(['Alice','Bob','Charlie','Diana','Edward'], 120),
    'Region': np.random.choice(['North','South','East','West'], 120),
    'Product': np.random.choice(['Laptop','Phone','Tablet','Monitor','Accessories'], 120),
    'Units_Sold': np.random.randint(1, 50, 120),
    'Unit_Price': np.random.choice([299, 499, 799, 1099, 49], 120),
}

df = pd.DataFrame(data)
df['Total_Sales'] = df['Units_Sold'] * df['Unit_Price']

print(df.head(10))
print(f"\nDataset shape: {df.shape}")
```

```python
# Total sales by salesperson
salesperson_performance = df.groupby('Salesperson')['Total_Sales'].sum().sort_values(ascending=False)

print("=== SALESPERSON PERFORMANCE ===")
print(salesperson_performance)
```

```python
# Total sales by region
region_performance = df.groupby('Region')['Total_Sales'].sum().sort_values(ascending=False)

print("=== REGIONAL PERFORMANCE ===")
print(region_performance)
```

```python
# Total sales by product
product_performance = df.groupby('Product')['Total_Sales'].sum().sort_values(ascending=False)

print("=== PRODUCT PERFORMANCE ===")
print(product_performance)
```

```python
# Visualizing salesperson performance
plt.figure(figsize=(10,5))
sns.barplot(x=salesperson_performance.index, y=salesperson_performance.values, palette='viridis')
plt.title('Total Sales by Salesperson')
plt.xlabel('Salesperson')
plt.ylabel('Total Sales ($)')
plt.tight_layout()
plt.show()
```

```python
# Region and Product charts side by side
fig, axes = plt.subplots(1, 2, figsize=(14,5))

# Region chart
sns.barplot(ax=axes[0], x=region_performance.index, y=region_performance.values, palette='magma')
axes[0].set_title('Total Sales by Region')
axes[0].set_xlabel('Region')
axes[0].set_ylabel('Total Sales ($)')

# Product chart
sns.barplot(ax=axes[1], x=product_performance.index, y=product_performance.values, palette='coolwarm')
axes[1].set_title('Total Sales by Product')
axes[1].set_xlabel('Product')
axes[1].set_ylabel('Total Sales ($)')

plt.tight_layout()
plt.show()
```
